```python
from tkinter import *
window=Tk()
window.title('Calculator')
window.geometry('400x400')

disp=Label(window,text='0',font=('bold'),height=3)
disp.grid(column=0,row=0)

window.s1=0
window.s2=0
window.a=0
window.b=0
window.c=0
window.d=1

def click1():
    click17()
    window.a=1
    window.b=1
    window.c=0
    window.d=1
bdiv=Button(window,text='/',height=2,width=5,font=('bold'),command=click1)
bdiv.grid(column=0,row=2)

def click2():
    click17()
    window.a=1
    window.b=2
    window.c=0
    window.d=1
    
bmulti=Button(window,text='*',height=2,width=5,font=('bold'),command=click2)
bmulti.grid(column=1,row=2)

def click3():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+1
        else:
            window.s1=window.s1+1/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+1
        else:
            window.s2=window.s2+1/(10**(window.d))
            window.d+=1
    display()
    
b1=Button(window,text='1',height=2,width=5,font=('bold'),command=click3)
b1.grid(column=0,row=3)

def click4():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+2
        else:
            window.s1=window.s1+2/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+2
        else:
            window.s2=window.s2+2/(10**(window.d))
            window.d+=1
    display()

b2=Button(window,text='2',height=2,width=5,font=('bold'),command=click4)
b2.grid(column=1,row=3)

def click5():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+3
        else:
            window.s1=window.s1+3/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+3
        else:
            window.s2=window.s2+3/(10**(window.d))
            window.d+=1
    display()
    
b3=Button(window,text='3',height=2,width=5,font=('bold'),command=click5)
b3.grid(column=2,row=3)

def click6():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+4
        else:
            window.s1=window.s1+4/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+4
        else:
            window.s2=window.s2+4/(10**(window.d))
            window.d+=1
    display()
    
b4=Button(window,text='4',height=2,width=5,font=('bold'),command=click6)
b4.grid(column=0,row=4)

def click7():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+5
        else:
            window.s1=window.s1+5/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+5
        else:
            window.s2=window.s2+5/(10**(window.d))
            window.d+=1
    display()
    
b5=Button(window,text='5',height=2,width=5,font=('bold'),command=click7)
b5.grid(column=1,row=4)

def click8():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+6
        else:
            window.s1=window.s1+6/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+6
        else:
            window.s2=window.s2+6/(10**(window.d))
            window.d+=1
    display()

b6=Button(window,text='6',height=2,width=5,font=('bold'),command=click8)
b6.grid(column=2,row=4)

def click9():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+7
        else:
            window.s1=window.s1+7/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+7
        else:
            window.s2=window.s2+7/(10**(window.d))
            window.d+=1
    display()

b7=Button(window,text='7',height=2,width=5,font=('bold'),command=click9)
b7.grid(column=0,row=5)

def click10():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+8
        else:
            window.s1=window.s1+8/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+8
        else:
            window.s2=window.s2+8/(10**(window.d))
            window.d+=1
    display()

b8=Button(window,text='8',height=2,width=5,font=('bold'),command=click10)
b8.grid(column=1,row=5)

def click11():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10+9
        else:
            window.s1=window.s1+9/(10**(window.d))
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10+9
        else:
            window.s2=window.s2+9/(10**(window.d))
            window.d+=1
    display()
    
b9=Button(window,text='9',height=2,width=5,font=('bold'),command=click11)
b9.grid(column=2,row=5)

def click12():
    if(window.a==0):
        if(window.c==0):
            window.s1=window.s1*10
        else:
            window.d+=1
    else:
        if(window.c==0):
            window.s2=window.s2*10
        else:
            window.d+=1
    display()
    
b0=Button(window,text='0',height=2,width=5,font=('bold'),command=click12)
b0.grid(column=2,row=6)

def click13():
    click17()
    window.a=1
    window.b=3
    window.c=0
    window.d=1
    
badd=Button(window,text='+',height=2,width=5,font=('bold'),command=click13)
badd.grid(column=2,row=2)

def click14():
    click17()
    window.a=1
    window.b=4
    window.c=0
    window.d=1
    
bmin=Button(window,text='-',height=2,width=5,font=('bold'),command=click14)
bmin.grid(column=3,row=2)

def click15():
    window.s1=0
    window.s2=0
    window.d=1
    window.c=0
    window.a=0
    window.b=0
    display()
    
bclear=Button(window,text='AC',height=2,width=5,font=('bold'),command=click15)
bclear.grid(column=3,row=6)

def click16():
    if(window.c==1):
        if(window.d==1):
            window.c=0
            if(window.a==0):
                window.s1=(window.s1*10)//10
            else:
                window.s2=(window.s2*10)//10
        else:
            if(window.a==0):
                window.s1=(window.s1*(10**(window.d-1)))//10
                window.d-=1
                window.s1=window.s1/(10**(window.d-1))
            else:
                window.s2=(window.s2*(10**(window.d-1)))//10
                window.d-=1
                window.s2=window.s2/(10**(window.d-1))
    else:
        if(window.a==0):
            window.s1=window.s1//10
        else:
            window.s2=window.s2//10
    display()

bdel=Button(window,text='C',height=2,width=5,font=('bold'),command=click16)
bdel.grid(column=0,row=6)

def click17():
    if(window.b==1):
        try:
            disp.configure(text=str(window.s1/window.s2))
            window.s1=window.s1/window.s2
        except ZeroDivisionError:
            disp.configure(text=str(0))
            window.s1=0
        window.b=0
        window.a=0      
        window.s2=0
    elif(window.b==2):
        disp.configure(text=str(window.s1*window.s2))
        window.b=0
        window.a=0
        window.s1=window.s1*window.s2
        window.s2=0
    elif(window.b==3):
        disp.configure(text=str(window.s1+window.s2))
        window.b=0
        window.a=0
        window.s1=window.s1+window.s2
        window.s2=0
    elif(window.b==4):
        disp.configure(text=str(window.s1-window.s2))
        window.b=0
        window.a=0
        window.s1=window.s1-window.s2
        window.s2=0
    else:
        window.a=0
    display()
    i=window.s1
    e=1
    while((i*10)%10!=0):
        e+=1
        i*=10
    if(e!=1):
        window.d=e
        window.c=1
        
beq=Button(window,text='=',height=2,width=5,font=('bold'),command=click17)
beq.grid(column=3,row=3)

def click18():
    window.c=1
    display()

bdec=Button(window,text='.',height=2,width=5,font=('bold'),command=click18)
bdec.grid(column=1,row=6)
    
def display():
    if(window.a==0):
        x=str(window.s1)
        disp.configure(text=x)
    else:
        x=str(window.s2)
        disp.configure(text=x)

window.mainloop()
```
